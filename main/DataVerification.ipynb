```python
import pandas as pd
```

```python
# Pref is the output, VDCerr is the input
data = pd.read_csv('../data/pi/3phaseFLTKp0.5Ki5Step0.1No.0.csv')
data
```

```python
data.set_index('Time').plot()
```

```python
data.columns = ['Time', 'O', 'I']
```

```python
time_step = data.Time[1] - data.Time[0]
```

```python
# Calculate Input differential value
dI_dt = (data.I.to_numpy()[2:] - data.I.to_numpy()[:-2]) / (2 * time_step)
# Calculate Output differential value
dO_dt = (data.O.to_numpy()[2:] - data.O.to_numpy()[:-2]) / (2 * time_step)
```

```python
pd.DataFrame([dI_dt*0.907+data.I.to_numpy()[1:-1]*0.00169, dO_dt]).T.plot()
```

```python
pd.DataFrame([dI_dt*0.5+data.I.to_numpy()[1:-1]*5 - dO_dt]).T.plot()
```
