```python
import pandas as pd
# import deepxde as dde
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
fnamevar = "PI_variables_0118.dat"
```

```python
lines = open(fnamevar, "r").readlines()
# read output data in fnamevar (this line is a long story...)
Chat = np.array(
    [
        np.fromstring(
            min(re.findall(re.escape("[") + "(.*?)" + re.escape("]"), line), key=len),
            sep=",",
        )
        for line in lines
    ]
)
l, c = Chat.shape
plt.plot(range(l), Chat[:, 0], "r-")
plt.plot(range(l), Chat[:, 1], "b-")
plt.plot(range(l), np.ones(Chat[:, 0].shape) * 0.5, "r--")
plt.plot(range(l), np.ones(Chat[:, 0].shape) * 5, "b--")
plt.legend(["Kp","Ki"], loc="right")
plt.xlabel("Epoch")

Chat[-10:-1]
```
