```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
fnamevar = "PI_variables_swish_transall_ObAll_0226.dat"
lines = open(fnamevar, "r").readlines()
# read output data in fnamevar (this line is a long story...)
Chat = np.array(
    [
        np.fromstring(
            min(re.findall(re.escape("[") + "(.*?)" + re.escape("]"), line), key=len),
            sep=",",
        )
        for line in lines
    ]
)
l, c = Chat.shape
plt.plot(range(l), Chat[:, 0], "r-")
plt.plot(range(l), Chat[:, 1], "b-")
plt.plot(range(l), Chat[:, 2], "g-")
plt.plot(range(l), Chat[:, 3], "y-")
plt.plot(range(l), np.ones(Chat[:, 0].shape) * 0.025, "r--")
plt.plot(range(l), np.ones(Chat[:, 0].shape) * 0.5, "r--")
plt.plot(range(l), np.ones(Chat[:, 0].shape) * 5., "r--")

plt.legend(["Kp1","Ki1", "Kp2","Ki2"], loc="right")
plt.xlabel("Epoch")
Chat[-10:-1]
```
