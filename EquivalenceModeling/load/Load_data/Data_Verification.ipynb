```python
import pandas as pd
```

```python
pd_data = pd.read_csv('0612_island_IM.csv', nrows=5000)
pd_data
```

```python
pd_data.columns = ['Time', 'theta', 'Id', 'Idrms', 'Iq', 'Iqrms', 'Ws','p_test', 'q_test', 'Tmpu', 'Vd', 'Vdrms', 'Vq', 'Vqrms', 'Ia', 'Ib', 'Ic',  'ph', 'qh', 'Wr', 'Te', 'Va', 'Vb', 'Vc']
```

```python
pd_data
```

```python
def v1(row):
    return (row.Vdrms * row.Idrms + row.Vqrms * row.Iqrms)*1.0

def v2(row):
    return (row.Vd * row.Id + row.Vq * row.Iq)
```

```python
pd_data['phCalcPeak'] = pd_data.apply(v2, axis=1)
pd_data['phCalcRMS'] = pd_data.apply(v1, axis=1)
```

```python
pd_data[['phCalcPeak','ph']].plot()
```

```python
(pd_data['ph']-pd_data['phCalcPeak']).describe()
```

```python
# Verification for reactive power qh
# RMS value
def v3(row):
    return (row.Vqrms * row.Idrms - row.Vdrms * row.Iqrms)*1.0
# peak value
def v4(row):
    return (row.Vq * row.Id - row.Vd * row.Iq)
pd_data['qhCalcPeak'] = pd_data.apply(v4, axis=1)
pd_data['qhCalcRMS'] = pd_data.apply(v3, axis=1)
pd_data[['qh','qhCalcPeak']].plot()
```

```python
rs = 2.0110e-2  # 定子电阻，单位为pu
xs = 1.0448e-1  # 定子电抗，单位为pu
xmu = 9.0424    # 磁化电抗，单位为pu
rr = 4.5768e-2  # 转子电阻，单位为pu
xr = 1.0448e-1  # 转子电抗，单位为pu
Omega_b = 377   
```

```python
x0 = xs + xmu
xp = xs + xr*xmu/(xr+xmu)
T0p = (xr+xmu)/(Omega_b * rr)
```

```python
pd_data['edp'] = pd_data.Vdrms - (rs * pd_data.Idrms - xp * pd_data.Iqrms)
pd_data['eqp'] = pd_data.Vqrms - (rs * pd_data.Iqrms + xp * pd_data.Idrms)
pd_data['edp'].plot()
# pd_data['eqp'].plot()
```

```python
edp_dt = list((pd_data.edp[2:].to_numpy() - pd_data.edp[:-2].to_numpy()) / (pd_data.Time[2] - pd_data.Time[0]))
eqp_dt = list((pd_data.eqp[2:].to_numpy() - pd_data.eqp[:-2].to_numpy()) / (pd_data.Time[2] - pd_data.Time[0]))
```

```python
pd_data['edp_dt'] = [edp_dt[0]] + edp_dt + [edp_dt[-1]]
pd_data['eqp_dt'] = [eqp_dt[0]] + eqp_dt + [eqp_dt[-1]]
pd_data['edp_dt'].plot()
# pd_data['eqp_dt'].plot()
```

```python
def edp_verification(row):
    return  (1 - row.Ws/377) * row.eqp * 377 - (row.edp + (x0 - xp) * row.Iqrms)/T0p

def eqp_verification(row):
    return  - (1 - row.Ws/377) * row.edp * 377 - (row.eqp - (x0 - xp) * row.Idrms)/T0p
```

```python
pd_data['edp_Calc'] = pd_data.apply(edp_verification, axis=1)
pd_data['eqp_Calc'] = pd_data.apply(eqp_verification, axis=1)
```

```python
pd_data[['edp_dt','edp_Calc']].plot()
(pd_data['edp_Calc']-pd_data['edp_dt']).describe()
pd_data['edp_Calc']
pd_data['edp_dt']
```

```python
pd_data[['eqp_Calc','eqp_dt']].plot()
(pd_data['eqp_Calc']-pd_data['eqp_dt']).describe()
```
