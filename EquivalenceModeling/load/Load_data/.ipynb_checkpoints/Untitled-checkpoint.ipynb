```python
pd_data.columns = ['TIme', 'Id', 'Idrms', 'Iq', 'Iqrms', 'Omega1', 'Omega2','p_test', 'q_test', 'Tmpu', 'Vd', 'Vdrms', 'Vq', 'Vqrms', 'ph', 'qh', 'sout', 'tau']
```

```python
pd_data
```

```python
def v1(row):
    return (row.Vd*row.Id + row.Vq*row.Iq)
```

```python
pd_data['qhv1'] = pd_data.apply(v1, axis=1)
```

```python
pd_data[['ph','qhv1']].plot()
```
