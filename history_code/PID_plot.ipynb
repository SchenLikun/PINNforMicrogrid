```python
import pandas as pd
```

```python
from pidv2 import *
```

```python
input_data_path = 'data//pi/3phaseFLTKp0.5Ki5Step0.1No.1.csv'
microgird = PID(path=input_data_path,
                init_Kp = 1.,
                init_Ki = 1.
                )
```

```python
microgird.input_data.plot()
```

```python
data = microgird.input_data
```

```python
time_step = data.index[1] - data.index[0]
```

```python
I_t = (data.I.to_numpy()[2:] - data.I.to_numpy()[:-2]) / (2 * time_step)
```

```python
O_t = (data.U.to_numpy()[2:] - data.U.to_numpy()[:-2]) / (2 * time_step)
```

```python
real_Kp = 0.5
real_Ki = 5
```

```python
pd.DataFrame([I_t*real_Kp+data.I.to_numpy()[1:-1]*real_Ki, O_t]).T.plot()
```

```python
K_p = 0
K_i = 0
```

```python
pd.DataFrame([I_t*K_p+data.I.to_numpy()[1:-1]*K_i, O_t]).T.plot()
```
